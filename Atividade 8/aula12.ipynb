```python
import matplotlib.pyplot as plt
from numpy import sin, cos #Para as contas devo usar em radianos.
```

```python
n = int(input('Insira um valor de n (valor maior que 1).'))
A = float(input('Insira o valor de da amplitude A da função.'))
tf = 10
dt = 0.1
t = [0]
x = [0]
```

Bn1 = A/2 (cos(n) / n + sin(n) / nˆ2)

```python
def Bn1(n, A):
    return (A*0.5) * ( (cos(n) / n) + (sin(n) / (n**2)) )
```

Bn2 = 1/2 \[-A (-cos(n) / n + sin(n) / nˆ2 + 3cos(3n) / n - sin(3n) / nˆ2) + 2A (-1/n * cos(n) + 1/3n cos(3n))\]

```python
def Bn2(n, A):
    return 0.5 * ( -A* ( (-cos(n)/n) + (sin(n)/(n**2)) + (3*cos(3*n)/n) - (sin(3*n)/(n**2))) + (2*A)* ((-1/n) * cos(n) + (1/(3*n)) * cos(3*n)))
```

Bn3 = 1/2 \[A ( -3cos(3n)/n + sin(3n)/nˆ2 + 4cos(4n)/n - sen(4n)/nˆ2) - 4A(-1/3n cos(3n) + 1/4n cos(4n))\]

```python
def Bn3(n, A):
    return 0.5 * ( A*( -(3*cos(3*n) / n) + (sin(3*n) / (n**2)) + (4*cos(4*n) / n) - (sin(4*n) / (n**2)) ) - 4*A*( -( 1/(3*n) * cos(3*n)) + (1/(4*n) *cos(4*n)) ) )
```

```python
def somatorio(N, A, t):
    n = 1
    s = 0
    while n <= N:
        if t % 4 == 1:
            s = s + Bn1(N, A) * sin(N*t)
        elif t % 4 <= 3:
            s = s + Bn2(N, A) * sin(N*t)
        else:
            s = s + Bn3(N, A) * sin(N*t)
        n += 1
    return s
```

```python
while t[-1] <= tf:
    t.append(t[-1] + dt)
    x.append(somatorio(n, A, t[-1]))
```

```python
plt.plot(t, x, label='x(t)')
```
