<p>Refatorando <b>a)</b>:</p>
<p>y[n] = x[n] +  1/2 y[n-1], x[n] = (1/3)ˆn e y[-1] = 1 </p>
<p>y[0] = x[0] + 1/2 y[0-1]</p>
<p>y[0] = 1 + 1/2 * 1 = 1,5</p>

```python
def verifica_entrada_num(str):
    entrada = input(str)
    if any(chr.isdigit() for chr in entrada):
        return int(entrada)
    print('Entrada inválida. Insira novamente um valor válido.')
    return verifica_entrada_num(str)
```

```python
def x (n):
    return (1/3)**n
def y1(y,n):
    return x(n) + 0.5 * y
```

```python
y = [1,1.5]
t = [-1,0]
T = 0.5
```

```python
n = verifica_entrada_num("Digite o limite superior de n [inteiro]")
```

```python
while t[-1] <= n:
    t.append(t[-1]+T)
    y.append( y1( y[-1] , t[-1]) )
```

```python
plt.plot(t, y, label='y(t)')
```
