Importando bibliotecas para apresentar os dados

```python
import sympy
sympy.init_printing()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Definindo variáveis para usar na função

```python
t, s = sympy.symbols('t, s')
m = sympy.symbols('m', real=True, positive=True)
c = sympy.symbols('c', real=True, positive=True)
k = sympy.symbols('k', real=True, positive=True)
f0 = sympy.symbols('f0', real=True, positive=True)
```

Funções para calcular as transforamadas

```python
def invL(F):
    return sympy.inverse_laplace_transform(F, s, t)
```

```python
def L(f):
    return sympy.laplace_transform(f, t, s, noconds=True)
```

Definindo as expressões para o cálculo

```python
y = sympy.Function('y')
u = sympy.Function('u')
y = m * y(t).diff(t).diff(t) + c * y(t).diff(t) + k * y(t) 
y
```

```python
Y = L(y) - f0/s
Y.subs({c: 2, f0: 3, k:1, m: 1})
```

```python
y = invL(Y)
y
```

```python
Ys =  sympy.Function('Ys')
Ys = f0/ (s * (m * s**2 + c * s + k))
Ys
```

```python
invL(Ys)
```

```python
p = sympy.plot(Ys.subs({m: 2, c: 3, k: 3, f0: 2}), Ys.subs({m: 2, c: 3, k: 3, f0: 2}),
               xlim=(0, 4), ylim=(0, 3), show=False)
p[1].line_color = 'red'
p.show()
```

```python
p = sympy.plot(invL(Ys).subs({m: 2, c: 3, k: 3, f0: 2}), invL(Ys).subs({m: 2, c: 3, k: 3, f0: 2}),
               xlim=(0, 4), ylim=(0, 3), show=False)
p[1].line_color = 'red'
p.show()
```
