```python
import numpy as np
import matplotlib.pyplot as plt
from math import e, pi, cos, sin
```

<h1>Exercício 1</h1>

```python
n1 = int(input('Insira um valor de n (valor maior que 1).'))
p = float(input('Insira um valor de p (valor marcado com ponto).'))
L = int(input('Insira um valor de L (o gráfico irá de 0 a L).'))
t = [-2]
x = [0]
A0 = 186.3098532236937
T0 = 4
W0 = pi/2
e8 = e**8
```

```python
def An1(n):
    return 0.5 * ( ( ( (n*W0*sin(2*n*W0))-(4*cos(2*n*W0)) ) / (e8 * ((n*W0)**2+16) ) ) - ( ( (-4*e8*cos(2*n*W0)) - (e8*n*W0*sin(2*n*W0)) ) / ((n*W0)**2+16) ) )
```

```python
def Bn1(n):
    return 0.5 * ( ( ( (-n*W0*cos(2*n*W0))-(4*sin(2*n*W0)) ) / (e8 * ((n*W0)**2+16) ) ) - ( ( (-e8*n*W0*cos(2*n*W0)) + (4*e8*sin(2*n*W0)) ) / ((n*W0)**2+16) ) )
```

```python
def somatorio1(N, x):
    n = 1
    s = 0
    while n <= N:
        s = s + An1(n) * cos(n*W0*x) + Bn1(n) * sin(n*W0*x)
        n += 1
    return s
```

```python
while t[-1] <= L:
    t.append(t[-1] + p)
    x.append(A0+somatorio1(n1, t[-1]))
```

```python
plt.plot(t, x, label='f(x)')
```
