```python
import matplotlib.pyplot as plt 
from math import cos, e, pi
```

```python
GtR = pi/180 #Grau para Radianos
```

```python
tempo = []
x1 = []
for t in range(0, 21):
    tempo.append(t/10)
    x1.append(t*10)
```

```python
def monta_func (A, alpha, omega, fi):
    """
    Função para calcular uma função senoidal com relação a var Tempo : List
    A é amplitude em Metros
    alpha é o fator de decaimento no tempo
    omega é a frequência em rad/s
    fi é a fase em graus

    Retornando um dicionário
    """
    x_t = []; e_t = []; osc_t =[]
    for tmp in tempo:
        x_t.append(A*(e**(-tmp*alpha))*cos(omega*t + fi*GtR))
        e_t.append(e**(-tmp*alpha))
        osc_t.append(A*cos(omega*tmp))
    return {'x_t' : x_t, 'e_t' : e_t, 'osc_t' : osc_t}
#print(exerc_a()['x_t'])
#print(len(exerc_a()['x_t']))
print(monta_func(3, 1, 9, 30))
#print(len(monta_func(3, 1, 9, 30)['x_t']))
```

```python
plt.title('Exercício A) graph - Gustavo R.')
funcao = monta_func(3, 1, 9, 30)

plt.plot(tempo,funcao['x_t'],label='x(t)')
plt.plot(tempo,funcao['e_t'],label='x(t)')
plt.plot(tempo,funcao['osc_t'],label='x(t)')
plt.xlabel('tempo (t)')
plt.ylabel('valores (x1)')

plt.legend()
plt.show()

print("A)")
```
