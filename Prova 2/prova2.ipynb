```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from math import e
```

<p>b) y + e<sup>y</sup> - x = 2</p> 
<p>dy/dx = 1 / (1 - e<sup>y</sup>)
<p>A derivada não pode ocorrer em zero</p>
<p>O resultado da função é em um plano tridmensional</p>

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
    
def funcao_z(x, y):
    return y + e**y - x - 2
    
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
    
X, Y = np.meshgrid(x, y)
Z = funcao_z(X, Y)
    
ax.plot_wireframe(X, Y, Z, color='green')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
    
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                    cmap='viridis', edgecolor='none')
ax.set_title('Gráfico de b');
    
plt.show()
```

```python
def derivada_b (y):
    if y !=0:
        return 1 / (1 - e**y)
y = [-6]
dydx = [derivada_b(y[-1])]
h = 0.1
while y[-1] < -0.5:
    y.append(y[-1]+h)
    dydx.append(derivada_b(y[-1]))
plt.subplot(1,2,1)
plt.plot(y, dydx)
y = [0.5]
dydx = [derivada_b(y[-1])]
while y[-1] < 6:
    y.append(y[-1]+h)
    dydx.append(derivada_b(y[-1]))
plt.subplot(1,2,2)
plt.plot(y, dydx)
```

<p>c) (x<sup>2</sup> + y<sup>2</sup>)<sup>3/2</sup> = 10xy</p> 
<p>dy/dx = (3x(x<sup>2</sup>  + y<sup>2</sup>) - 10y) / (10x -3y(x<sup>2</sup>  + y<sup>2</sup>))</p>
<p>Os resultados da função e da derivada são dados em um plano tridmensional</p>

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
    
def funcao_z(x, y):
    return (x**2 + y**2)**(3/2) - 10*x*y
    
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
    
X, Y = np.meshgrid(x, y)
Z = funcao_z(X, Y)

ax.plot_wireframe(X, Y, Z, color='green')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
    
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                    cmap='viridis', edgecolor='none')
ax.set_title('Gráfico de c');
    
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

def derivada(x, y):
    return ( (3*x*( (x**2 + y**2)**(1/2)) ) - (10*y) ) / ( (10*x)-( 3*y*( (x**2 + y**2)**(1/2)) ) )
    
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
    
X, Y = np.meshgrid(x, y)
Z = derivada(X, Y)

ax.plot_wireframe(X, Y, Z, color='green')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
    
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                    cmap='viridis', edgecolor='none')
ax.set_title('Gráfico da derivada de c');
    
plt.show()
```
